```python
def nat_read_xlsx(xrow1,xrow2,xcol):
    import xlrd
    file = 'RegionalTotals.xlsx'
    wb=xlrd.open_workbook(file)
    sheet=wb.sheet_by_index(0)
    data=[]
    for i in range (xrow1,xrow2):
        data.append(sheet.cell_value(i,xcol))
    return data
```
